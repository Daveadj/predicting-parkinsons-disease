```python
from sklearn.preprocessing import LabelEncoder
```

```python
lc=LabelEncoder()
```

```python
df['name']=lc.fit_transform(df['name'])
```

```python
df.columns
```

```python
#reshaping the columns in the dataset by putting the target variable/column last
reshape_df=df[['name', 'MDVP:Fo(Hz)', 'MDVP:Fhi(Hz)', 'MDVP:Flo(Hz)', 'MDVP:Jitter(%)',
       'MDVP:Jitter(Abs)', 'MDVP:RAP', 'MDVP:PPQ', 'Jitter:DDP',
       'MDVP:Shimmer', 'MDVP:Shimmer(dB)', 'Shimmer:APQ3', 'Shimmer:APQ5',
       'MDVP:APQ', 'Shimmer:DDA', 'NHR', 'HNR', 'RPDE', 'DFA',
       'spread1', 'spread2', 'D2', 'PPE','status']]
```

```python
reshape_df
```

# split the dataset

```python
x=reshape_df.drop('status',axis=1)
y=reshape_df['status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

# min max scaler

```python
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
x_train=ms.fit_transform(x_train)
x_test=ms.transform(x_test)
```

# random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier(n_estimators=10,criterion='entropy')
rfc.fit(x_train,y_train)
```

```python
prediction=rfc.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,prediction)
print(cm)
accuracy_score(y_test,prediction)*100
```

# sector vector machine

```python
from sklearn.svm import SVC
```

```python
classify=SVC(kernel='linear')
classify.fit(x_train,y_train)
```

```python
prediction_2=classify.predict(x_test)
```

```python
cm2=confusion_matrix(y_test,prediction_2)
print(cm2)
accuracy_score(y_test,prediction_2)*100
```

# decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc=DecisionTreeClassifier(criterion='entropy')
dtc.fit(x_train,y_train)
```

```python
prediction_3=dtc.predict(x_test)
```

```python
cm_3=confusion_matrix(y_test,prediction_3)
print(cm_3)
accuracy_score(y_test,prediction_3)*100
```

# logistic regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
prediction_4=lr.predict(x_test)
```

```python
cm_4=confusion_matrix(y_test,prediction_4)
print(cm_4)
accuracy_score(y_test,prediction_4)*100
```

# xgboost classifier

```python
from xgboost import XGBClassifier
```

```python
xgb_model=XGBClassifier()
xgb_model.fit(x_train,y_train)
```

```python
prediction_5=xgb_model.predict(x_test)
```

```python
cm_5=confusion_matrix(y_test,prediction_5)
print(cm_5)
accuracy_score(y_test,prediction_5)*100
```
